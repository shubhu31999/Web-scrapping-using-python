## IMDB Web Scraping

## Importing all the necessary liaberaries.

```python
import requests
import openpyxl
from bs4 import BeautifulSoup
url="https://www.imdb.com/chart/top/?ref_=nv_mv_250"
```

##  Step 1- Get the HTML

```python
source=requests.get(url)
source.raise_for_status()
```

```python
print(source)
```

## step 2 - Parse the HTML

```python
soup=BeautifulSoup(source.content,"html.parser")
```

## Step 3- Creating a Excel workbook from openpyxl

```python
excel=openpyxl.Workbook()
print(excel.sheetnames)
sheet=excel.active
sheet.title="IMDB top movies"
name=sheet.title
print("the sheet name is:",name)
sheet.append(['Rank','Movie name','Year Of Release','IMDB Ratings'])
```

## Find all paras and iterating through them.

```python
movies=soup.find('tbody',class_='lister-list').find_all('tr')



for movie in movies:
    name=movie.find('td', class_="titleColumn").a.text
    rank=movie.find('td', class_="titleColumn").text.split('.')[0]
    year=movie.find('span',class_='secondaryInfo').text.strip('()')
    ratings=movie.find('td',class_='ratingColumn imdbRating').strong.text
    print(rank,name,year,ratings)
    sheet.append([rank,name,year,ratings])
    
excel.save('IDMB top Rated Movies.xlsx')
```
